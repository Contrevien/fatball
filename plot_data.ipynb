```python
import numpy as np
data = np.loadtxt("./test_data.csv", delimiter=",")
```

```python
train_features = data[:, :3]
train_targets = data[:, 3]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(train_features[:, 0], train_features[:, 1], c=train_targets, s=30)
legend1 = ax.legend(*scatter.legend_elements(),loc="lower left", title="Wait(0) Move(1)")
ax.add_artist(legend1)
plt.xlabel("Temperature")
plt.ylabel("Food")
plt.show()
```
